# 1. Tạo Dataframe như dữ liệu đã cho trên.
# 2. Hiển thị thông tin Dataframe vừa tạo.

```python
import pandas as pd 

data={
    "Name":["Mr. Harris","Mrs. Bradley","Miss. Laina","Mrs. Heath","Mr. William","Mr. James","Mr. Timothy","Mrs. Oscar","Mrs. Nicholas","Miss. Marguerite","Mr. William","Miss. Hulda","Mr. Lawrence","Mr. Wilam"],
   "Height":[1.80,1.61,1.78,1.74,1.55,1.9,1.61,1.40,1.92,1.77,1.55,1.85,1.64,1.77],
    "Weight": [90,87,64,70,100,106,82,55,77,60,100,109,80,66],
    "AGE":[32,45,22,52,48,21,28,69,58,21,48,33,25,70]
}
df = pd.DataFrame(data)
print(df)
```

# 3. Tính giá trị trung bình của cột "AGE".

```python
mean_age = round(df['AGE'].mean())
mean_age
```

# 4. Tính trung bình cột weight 

```python
mean_weight = round(df['Weight'].mean())
mean_weight
```

# 5. Sấp xếp Dataframe theo cột Weight tăng dần.

```python
df_weight = df[['Weight']].sort_values(by='Weight', ascending=True)
print("Sắp xếp dữ liệu điểm cột weight tăng dần:")
print(df_weight)
```

# 6. Vẽ biểu đồ cột cho AGE.

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(20, 5))
plt.bar(df['Name'], df['AGE'], color ="yellow")
plt.title("Biểu đồ cột")
plt.ylabel("Name")
plt.xlabel("AGE")
plt.show()
```

# 7. Vẽ biểu đồ phân tán cho AGE và Height.

```python
# Biểu đồ phân tán

plt.scatter(df['AGE'], df['Height'], alpha = 0.5, color ="pink")
plt.title("Biểu đồ phân tán")
plt.xlabel("AGE")
plt.xlabel("Height")
plt.show()
```

# 8. Vẽ biểu đồ đường cho cột Weight.

```python
plt.plot(df['Weight'], label ="Biểu đồ đường", color ="red",marker ="o")
plt.title("Biểu đồ đường")
plt.xlabel("Trục x")
plt.xlabel("Trục y")
plt.legend()
plt.show()
```
