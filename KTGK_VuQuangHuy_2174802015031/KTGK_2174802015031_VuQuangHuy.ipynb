# 14. Biết công thức tính chỉ số BMI như sau:
# BMI = Weight(kg)/Height2 (m2)
# Tạo một cột mới trong Dataframe với tên cs_BMI, tính chỉ số BIM mỗi người

```python
data={
    "Name":["Mr. Harris","Mrs. Bradley","Miss. Laina","Mrs. Heath","Mr. William","Mr. James","Mr. Timothy","Mrs. Oscar","Mrs. Nicholas","Miss. Marguerite","Mr. William","Miss. Hulda","Mr. Lawrence","Mr. Wilam"],
   "Height":[1.80,1.61,1.78,1.74,1.55,1.9,1.61,1.40,1.92,1.77,1.55,1.85,1.64,1.77],
    "Weight": [90,87,64,70,100,106,82,55,77,60,100,109,80,66],
    "AGE":[32,45,22,52,48,21,28,69,58,21,48,33,25,70]

}
df = pd.DataFrame(data)
df['cs_BMI'] = df['Weight'] / (df['Height'] ** 2)
print(df)
```

# 15. Theo thang đo kết quả tính BMI của châu Á:
# Chỉ số BMI dưới 18,5 là thiếu cân, 18.5-22,9 là bình thường, 23-24,9 là thừa cân, từ 25 trở lên là béo phì.
# Tạo cột với tên 'Phan_Loai' với các mức ‘0’,’1’,’2’,’3’ tương ứng với mức chỉ số BMI trên.

```python
def classify_bmi(cs_BMI):
    if cs_BMI < 18.5:
        return 0  
    elif 18.5 <= cs_BMI <= 22.9:
        return 1  
    elif 23 <= cs_BMI <= 24.9:
        return 2  
    else:
        return 3  

df['Phan_Loai'] = df['cs_BMI'].apply(classify_bmi)
df
```

# 16. Chuẩn hóa giá trị của cột cs_BMI chuẩn hóa giá trị trong đoạn từ 0 tới 1.

```python
df['cs_BMI_normalized'] = (df['cs_BMI'] - df['cs_BMI'].min()) / (df['cs_BMI'].max() - df['cs_BMI'].min())
df
```

## 17. Cho biết:
+ ‘Mr‘tương ứng với ‘male’,
+ ’Mrs’ và ‘Miss’ tương ứng với ‘female’.
Tạo cột với tên GT, Sau khi thực hiện, xóa các từ Mr, Mrs, Miss nếu trong
cột Name còn các từ đó.

```python
def extract_title(df):
   
    df['GT'] = df['Name'].str.extract(r'(\bMr\b|\bMrs\b|\bMiss\b)', expand=False)
    return df


df = extract_title(df)
print('đã chuyển đổi sang mr,ms')
df
def substitute_Sex(x_df):
    mapping={'Mr':'male', 'Mrs':'female', 'Miss':'female'}
    x_df['GT']=x_df['GT'].map(mapping)
    return x_df

tidisater5=df.pipe(substitute_Sex)
tidisater5
```
